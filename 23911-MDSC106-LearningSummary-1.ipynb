Creating Random number to make a adhar numbers list

```python
import random

def generate_unique_numbers(count, length):
    numbers = set()
    while len(numbers)<count:
        num = str(random.randint(10**(length-1), 10**length - 1)).zfill(length)
        numbers.add(num)
    return numbers

unique_numbers = generate_unique_numbers(442, 12)

adhar=[int(i) for i in unique_numbers]
l=random.randint(10,30)
random_list = [random.randint(0, 441) for _ in range(l)]
for _ in random_list:
    adhar[_]=None
```

Function to plot scatter, bar and histplot for given df and col and target.

```python
def plot(df,col,target):
    plt.figure(figsize=(10,5))

    plt.subplot(1,3,1)
    plt.hist(df[col],bins=50,edgecolor='black')
    plt.title('Histogram')

    plt.subplot(1,3,2)
    plt.scatter(df[col],df[target],color='r')
    plt.title('Scatter')

    plt.subplot(1,3,3)
    sns.barplot(x=target,y=col,data=df,color='g')
    plt.title('Bar Plot')

    plt.grid()
    plt.show()
```

Calling the plot function for diff columns

```python
for col in df:
    if col!='s1':
        plot(df,col,'s1')
```

#**Histogram**

```python
def histplot(df,col):
    class1=df[df['class']==1]
    class2=df[df['class']==2]
    class3=df[df['class']==3]

    fig,axes = plt.subplots(1,3,figsize=(15,5))
    fig.suptitle(col,fontsize=20)

    axes[0].set_title('Class-1',fontsize=10)
    axes[1].set_title('Class-2',fontsize=10)
    axes[2].set_title('Class-3',fontsize=10)

    sns.histplot(data=class1,x=col,bins=20, kde=True,ax=axes[0],color='r')
    sns.histplot(data=class2,x=col,bins=20, kde=True,ax=axes[1],color='g')
    sns.histplot(data=class3,x=col,bins=20, kde=True,ax=axes[2],color='b')

    plt.show()
```

```python
for col in df.columns:
    if col!='class':
        histplot(df,col)
```
